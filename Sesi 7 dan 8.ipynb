```python
df_total = df_can[years].sum(axis=0).to_frame()
df_total.index = df_total.index.astype(float) #more useful for regression as float
df_total.reset_index(inplace=True)
df_total.columns = ['year','total']
df_total
```

```python
ax = sns.regplot(x='year',y='total',data=df_total)
ax.set_title('Regreesion Plot Total Immigrants 1980-2010')
```

```python
#Optional apabila ingin memunculkan formula dari regresi linier
from scipy import stats

slope,intercept,r_value,p_value,std_err = stats.linregress(df_total['year'],df_total['total']) 
ax = sns.regplot(x='year',y='total',data=df_total,line_kws={'label':"y={0:.1f}x{1:.1f}".format(slope,intercept)}) #optional
ax.legend()
plt.show()
```

```python
#Change Color
ax = sns.regplot(x='year',y='total',data=df_total,color='blue')
ax.set_title('Regreesion Plot Total Immigrants 1980-2010')
```

```python
#Change marker
ax = sns.regplot(x='year',y='total',data=df_total,color='blue',marker='+')
ax.set_title('Regreesion Plot Total Immigrants 1980-2010')
```

```python
df_incidents=pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Police_Department_Incidents_-_Previous_Year__2016_.csv')
```

```python
df_incidents.head()
```

```python
#Mapping 100 row pertama dari tabel insiden
df_incidents_100 = df_incidents.iloc[0:100,:]
df_incidents_100
```

```python
#San Fransisco Lat and Lon
latitude= 37.77
longitude=-122.42
```

```python
sf_map = folium.Map(location=[latitude,longitude],zoom_start=12)
sf_map
```

```python
df_incidents_100
```

```python
sf_map
```

```python
# Clustering semua data sesuai kemiripan
from folium import plugins
sf_map = folium.Map(location=[latitude,longitude],zoom_start=12)

incidents = plugins.MarkerCluster().add_to(sf_map)

for lat,lon,label in zip(df_incidents_100['Y'],df_incidents_100['X'],df_incidents_100.Category):
    incidents.add_child(
        folium.CircleMarker(
            [lat,lon],
            radius=5,
            color='blue',
            fill=True,
            fill_color='red',
            fill_opacity=0.6,
            popup=label
        
        )
    
    )
sf_map.add_child(incidents) 
```
