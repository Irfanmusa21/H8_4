Rumus fungsi data berdistribusi uniform sebagai berikut.

$f(x) = \begin{cases}
\frac{1}{b-a}, \hspace{0.4cm} a \leq x \leq b \\
0, \hspace{0.7cm} other
\end{cases}$ <br>
Interval ditentukan berdasarkan argumen loc dan scalenya, sedangkan argumen size untuk menjelaskan jumlah variasi acak.

```python
data_uniform = uniform.rvs(size=10000, loc=10, scale=10)
sns.distplot(data_uniform)
plt.xlabel('Uniform Distribution')
plt.show()
```

Rumus fungsi data berdistribusi normal sebagai berikut. <br>
$f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}}$ <br>
<br>
Argumen loc sesuai dengan mean $(\mu)$ distribusi, scale sesuai dengan standard deviasi $(\sigma)$, dan size jumlah variasi acak

```python
from scipy.stats import norm

data_normal = norm.rvs(size=10000,loc=0,scale=1)

sns.distplot(data_normal)
```

Rumus fungsi data berdistribusi gamma sebagai berikut. <br>
$f(x) = \frac{x^{a-1}}{\beta^{a}\Gamma(a)}e^{-\frac{x}{\beta}}$ <br>
<br>
$\Gamma(a) = (a-1)!$ merupakan fungsi gamma <br>
Apabila a = 1, menghasilkan fungsi berdistribusi eksponensial

```python
from scipy.stats import gamma
data_gamma = gamma.rvs(size=10000,a=1) #a puncak
sns.distplot(data_gamma)
```

```python
from scipy.stats import expon
data_expon = expon.rvs(scale=1,loc=0,size=1000)
sns.distplot(data_expon)
```

Rumus probabilitas data berdistribusi poisson sebagai berikut. <br>
$p(x) = P(X=x) = \frac{e^{-\lambda}\lambda^{x}}{x!}$ <br>
<br>
X = 0,1,2... <br>
$\mu = \lambda$ <br>
$\sigma = \sqrt{\lambda}$

```python
from scipy.stats import poisson
data_poisson = poisson.rvs(mu=3,size=10000)
sns.distplot(data_poisson)
plt.xlabel('Poisson Distribution (Count Data)')
```

Rumus peluang data berdistribusi binomial sebagai berikut. <br>
$f(x) = \frac{n!}{(n-x)!}p^{x}q^{n-x}$ <br>
<br>
dengan <br>
n = jumlah percobaan <br>
p = peluang sukses <br>
q = peluang gagal

```python
from scipy.stats import binom
data_binom = binom.rvs(n=10,p=0.8,size=10000)
sns.distplot(data_binom)
```

```python
from scipy.stats import bernoulli
data_bern = bernoulli.rvs(size=10000,p=0.6)
sns.distplot(data_bern)
```

Confidence interval = populasi proporsi atau$ \mu \pm z -$ score*standard eror 

Standa eror (proporsi populasi) = $\sqrt{\frac{populasi proporsi (1-proporsi populasi)}{number observasi}}$ <br>
Standa eror ($\mu$) = $\frac{standar deviasi)}{\sqrt{number observasi}}$

```python
import numpy as np
import pandas as pd
df = pd.read_csv('heart.csv')
df.head()
```

```python
def sex_label(x):
    if x==1:
        return 'Male'
    else:
        return 'Female'
```

```python
#Standard eror dari proporsi populasi
n = 86+226
pro_fm = 226/n
print(n)
print(pro_fm)
stderror_fm = (pro_fm*(1-pro_fm)/n)**(1/2)
print(stderror_fm)
```

Confidence interval = populasi proporsi atau$ \mu \pm z-score$ *standard eror 

```python
#Misal confidence interval 95%, Z-score=1.96
z_score = 1.96
low_ci_fm = pro_fm - (z_score*stderror_fm)
up_ci_fm = pro_fm + (z_score*stderror_fm)
print(low_ci_fm)
print(up_ci_fm)
```

```python
print('Margin Error=',(z_score*stderror_fm))
```

```python
import scipy.stats as stats
import random
import warnings
```

```python
normal_distribution_male = norm.rvs(loc=181,scale=24,size=6000)
normal_distribution_fem = norm.rvs(loc=132,scale=22,size=6500)
all_mass = np.concatenate([normal_distribution_male,normal_distribution_fem],axis=0)
df_mass = pd.DataFrame(all_mass,columns=['people_mass'])
df_mass
```

```python
warnings.filterwarnings('ignore')
```

```python
sns.distplot(df_mass['people_mass'])
plt.ylabel('Probability of Occurence')
plt.xlabel('Mass (Pounds)')
plt.show()
```

```python
#Mean
population_mean = df_mass['people_mass'].mean()
population_mean
```

```python
# Standard Deviation
population_std = df_mass['people_mass'].std(ddof=1)
population_std
```

```python
#Mengambil 250 sampel masing-masing sejumlah 30 orang/sampel
n = 30
sample_means=[]
for sample in range(0,250):
    sample_values= df_mass['people_mass'].sample(n)
    sample_mean = sample_values.mean()
    sample_means.append(sample_mean)
sns.distplot(sample_means)
```

```python
df_mass['people_mass'].sample(10) #.sample()ambil sample random
```

```python
mean_of_sample_means = np.mean(sample_means)
median_of_sample_means = np.median(sample_means)
mean_of_sample_means , median_of_sample_means
```

```python
population_mean
```

```python
std_of_sample_means = np.std(sample_means)
std_of_sample_means
```

```python
standard_error = population_std/np.sqrt(n)
standard_error
```

```python
std_of_sample_means
```

Distribusi of sample means = $\bar{x}$ <br>
around population mean = $\mu$

confidence interval = $\bar{x} \pm z\frac{s}{\sqrt{n}}$

```python
stats.norm.cdf(2.17)
```

```python
z = 1.96
lower_bound = mean_of_sample_means - z * standard_error
lower_bound
```

```python
z = 1.96
upper_bound = mean_of_sample_means + z * standard_error
upper_bound
```

```python
kde = stats.gaussian_kde(sample_means)
pos = np.linspace(np.min(sample_means),np.max(sample_means),10000)
plt.plot(pos,kde(pos))
shade = np.linspace(lower_bound,upper_bound,300)
plt.fill_between(shade,kde(shade),alpha=0.5)
plt.xlabel('Sample Mass (Pounds)')
plt.ylabel('Frequency')
plt.axvline(x=mean_of_sample_means,color='red')
plt.text(x=152,y=.01,s='Shaded area are \npart of the \nconfidence intervals \n95%')
```

Uji hipotesis <br>
$\alpha =$ 1 - confidence level <br>
Jika p-value > $\alpha$ : hoptesis awal diterima <br>
p-value  $\leq \alpha$ : hoptesis awal ditolak

```python
df = pd.read_csv('nhanes_2015_2016.csv')
df
```

Uji apakah rata2 BMI dari Laki-Laki lebih tinggi dari pada rata2 BMI Perempuan <br>

$h_{0}$ : meanL $\leq$ meanP <br>
$h_{1}$ : meanL > meanP

```python
females = df[df.RIAGENDR==2]
males = df[df.RIAGENDR==1]
```

```python
# jumlah data, mean, stdeviasi
n_females = len(females)
mean_females = females['BMXBMI'].mean()
std_females = females['BMXBMI'].std()
print(n_females)
print(mean_females)
print(std_females)
```

```python
# jumlah data, mean, stdeviasi
n_males = len(males)
mean_males = males['BMXBMI'].mean()
std_males = males['BMXBMI'].std()
print(n_males)
print(mean_males)
print(std_males)
```

Diperhatikan bahwa mean L = 28.778072111846985 sedangkan mean P = 29.939945652173996. <br> Artinya, meanL $\leq$ meanP, $h_{0}$ diterima.

```python
import statsmodels.api as sm
```

```python
alpha = 0.05
bmi_hypo = sm.stats.ztest(males['BMXBMI'].dropna(),females['BMXBMI'].dropna(),alternative='larger')
bmi_hypo
```

```python
if bmi_hypo[1]>alpha:
    print('Accept Null Hypothesis (h0) Mean of Males BMI is not Larger than Females')
else:
    print('Reject Null Hypothesis (h0)  Mean of Males BMI is Larger than Females')
```

Uji apakah mean BMI Laki Laki tidak sama dengan mean BMI Perempuan <br>
$h_{0}$ : meanL = meanP <br>
$h_{1}$ : meanL $\neq$ meanP

```python
bmi_hypo_2s = sm.stats.ztest(males['BMXBMI'].dropna(),females['BMXBMI'].dropna(),alternative='two-sided')
bmi_hypo_2s
```

```python
if bmi_hypo_2s[1]>alpha:
    print('Accept Null Hypothesis (h0)')
else:
    print('Reject Null Hypothesis (h0)')
```

```python
n = 1018
pro_null = 0.52
pro_alt = 0.56
pro_hypo = sm.stats.proportions_ztest(n*pro_alt,n,pro_null,alternative='larger')
pro_hypo
```

```python
if pro_hypo[1]>alpha:
    print('Accept Null Hypothesis (h0)')
else:
    print('Reject Null Hypothesis (h0)')
```
